```python
import pandas as pd
```

## copy data from Excel

```python
df=pd.read_clipboard()
df
```

```python
df=pd.read_clipboard(names=['empId', 'firstname', 'lastname', 'gender'])
df
```

```python
df=pd.read_clipboard(index_col=0, names=['empId', 'firstname', 'lastname', 'gender'])
df
```

## copy data from CSV (Comma-Separated Values)

```python
df=pd.read_clipboard(sep=',')
df
```

## copy data from TSV (Tab-Separated Values)

```python
df=pd.read_clipboard(sep='\t')
df
```

## copy data from HTML Table

```python
# https://finance.yahoo.com/quote/SCC.BK/history?p=SCC.BK
# didn't work on Microsoft Edge browser
df=pd.read_clipboard()
df
```

```python
# https://finance.yahoo.com/quote/SCC.BK/history?p=SCC.BK
# work with Chrome/Firefox
df=pd.read_clipboard(index_col=0, parse_dates=True)
df
```

```python
%matplotlib inline
df['Close*'].plot()
```
