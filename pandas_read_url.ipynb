```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from datetime import datetime
print(datetime.now())
```

```python
url="https://en.wikipedia.org/wiki/Government_debt"
```

```python
dfs=pd.read_html(url)
dfs[2]
```

```python
dfs=pd.read_html(url,match="country")
dfs
```

```python
dfs=pd.read_html(url,match="country",header=0)
df=dfs[0]
df
```

```python
df.info()
```

```python
df['% of GDP']=df['% of GDP'].str.replace(r'%','').astype(float)
df["% of world public debt"]=df["% of world public debt"].str.replace(r"%","").astype(float)
df
```

```python
df.country=df.country.str.replace(r"[^\s+\w]","")
```

```python
plt.figure(figsize=(7,7))
sns.barplot(x="% of GDP",y="country",data=df.sort_values("% of GDP",ascending=False))
```

```python
dg=df.sort_values("% of GDP",ascending=False)
```

```python
color=["Red" if c == "Italy" else ".75" for c in dg["country"]]
```

```python
plt.figure(figsize=(7,7))
sns.barplot(x="% of GDP",y="country",data=dg,palette=color)
```
