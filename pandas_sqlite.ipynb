```python
import sqlite3
import pandas as pd
```

```python
db="chinook.sqlite"
con=sqlite3.connect(db)
```

```python
sql="""SELECT * FROM sqlite_master WHERE type='table'"""
df=pd.read_sql(sql,con)
df
```

```python
sql="SELECT * FROM Album limit 10"
df=pd.read_sql(sql,con)
df
```

```python
sql="""select * from Playlist limit 10"""
pd.read_sql(sql,con)
```

```python
sql="""select AlbumId, Title, a.ArtistId, b.Name 
        from Album a inner join Artist b on a.ArtistId = b.ArtistId 
        limit 10"""
pd.read_sql(sql,con)
```

```python
sql="""select AlbumId, Title, a.ArtistId, b.Name 
        from Album a inner join Artist b on a.ArtistId = b.ArtistId
        where title like ?
        limit 10"""
pd.read_sql(sql,con,params=['%Rock%'])
```
